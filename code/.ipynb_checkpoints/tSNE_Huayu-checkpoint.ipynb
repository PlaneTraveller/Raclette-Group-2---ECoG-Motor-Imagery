## Trajectory of all electrodes, single subject, default frequency bands

```python

from scipy.stats import zscore
import matplotlib as mpl

import torch
from torch import nn, optim
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE


def showTraj(subject = 0, trial = 0, task = 11):
  """ Shows trajectory of all electrodes in one patient """
  dat = alldat[subject][trial]
  V = dat['V'].astype('float32')


  # high-pass filter above 50 Hz
  b, a = signal.butter(3, [50], btype='high', fs=1000)
  V = signal.filtfilt(b, a, V, 0)

  # compute smooth envelope of this signal = approx power
  V = np.abs(V)**2
  b, a = signal.butter(3, [10], btype='low', fs=1000)
  V = signal.filtfilt(b, a, V, 0)

  # normalize each channel so its mean power is 1
  V = V/V.mean(0)

  # Plotting
  n_col = 1
  fig, ax = plt.subplots(1, n_col, figsize=(4.5, 4.5))

  # Extracting trials
  nt, nchan = V.shape
  nstim = len(dat['t_on'])

  trange = np.arange(0, 2000)
  ts = dat['t_on'][:, np.newaxis] + trange
  V_epochs = np.reshape(V[ts, :], (nstim, 2000, nchan))

  V_task = (V_epochs[dat['stim_id'] == task]).mean(0)

  Vtensor = torch.from_numpy(V_task)

  # First do PCA to reduce dimensionality to 20 dimensions so that tSNE is faster
  resp_lowd = PCA(n_components=min(15, Vtensor.shape[1]), random_state=0).fit_transform(Vtensor)

  # Then do tSNE to reduce dimensionality to 2 dimensions
  resp_lowd = TSNE(n_components=2, random_state=0).fit_transform(resp_lowd)
  print(resp_lowd.shape)
  # Plot dimensionality-reduced population responses 'resp_lowd'
  # on 2D axes, with each point colored by stimulus orientation
  time = np.arange(0, 2000)
  time = torch.from_numpy(time)
  print(time.shape)

  x, y = resp_lowd[:, 0], resp_lowd[:, 1]
  pts = ax.scatter(x, y, c=time, cmap='twilight', vmin=0, vmax=2000)
  fig.colorbar(pts, ax=ax, ticks=np.linspace(0, 2000, 1), label='time')

  ax.set_title("Subject: %s, Trial: %s, Task: %s" % (subject, trial, task))
  ax.set_xlabel('Dimension 1')
  ax.set_ylabel('Dimension 2')
  ax.set_xticks([])
  ax.set_yticks([])

showTraj()
showTraj(0, 1)

showTraj(0, 0, 12)
showTraj(0, 1, 12)

showTraj(2, 0)
showTraj(2, 1)

showTraj(2, 0, 12)
showTraj(2, 1, 12)
```

## Trajectory of electrodes in our ROI (Pre/post-central gyrus) from all patients, in different frequency bands

```python

from scipy.stats import zscore
import matplotlib as mpl

import torch
from torch import nn, optim
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

def roiTraj(task = 11, low = 10, high = 50):
  """ Shows the trajectory of all electrodes located in Precentral
   Gyrus and Postcentral Gyrus in all subjects.
  """

  roi1 = "Precentral Gyrus"
  roi2 = "Postcentral Gyrus"
  
  # Plotting
  n_col = 2
  fig, axs = plt.subplots(1, n_col, figsize=(9, 4.5))

  for i in range(2):
    
    ax = axs[i]
    if i == 0:
      ax.set_title("Movement")
    else:
      ax.set_title("Imagery")

      
    V_task = np.empty([2000, 1])
  
    for sub in range(7):
      dat = alldat[sub][i]
  
      V = dat['V'].astype('float32')
  
      # high-pass filter above 50 Hz
      b, a = signal.butter(3, [high], btype='high', fs=1000)
      V = signal.filtfilt(b, a, V, 0)
  
      # compute smooth envelope of this signal = approx power
      V = np.abs(V)**2
      b, a = signal.butter(3, [low], btype='low', fs=1000)
      V = signal.filtfilt(b, a, V, 0)
  
      # normalize each channel so its mean power is 1
      V = V/V.mean(0)
  
      # Extracting trials
      nt, nchan = V.shape
      nstim = len(dat['t_on'])
      trange = np.arange(0, 2000)
      ts = dat['t_on'][:, np.newaxis] + trange
      V_epochs = np.reshape(V[ts, :], (nstim, 2000, nchan))
      gyruses = np.array(dat["gyrus"])
      ind = (gyruses == roi1) | (gyruses == roi2)
      V_epochs = V_epochs[:, :, ind]
      V_new = V_epochs[dat['stim_id'] == task].mean(0)
      V_task = np.concatenate((V_task, V_new), axis=-1)
  
    V_task = np.nan_to_num(V_task)
    V_task = np.delete(V_task, 0, 1)
  
    Vtensor = torch.from_numpy(V_task)
  
    # First do PCA to reduce dimensionality to 20 dimensions so that tSNE is faster
    resp_lowd = PCA(n_components=min(20, Vtensor.shape[1]), random_state=0).fit_transform(Vtensor)
  
    # Then do tSNE to reduce dimensionality to 2 dimensions
    resp_lowd = TSNE(n_components=2, random_state=0).fit_transform(resp_lowd)
  
    # Plot dimensionality-reduced population responses 'resp_lowd'
    # on 2D axes, with each point colored by stimulus orientation
    time = np.arange(0, 2000)
    time = torch.from_numpy(time)
  
    x, y = resp_lowd[:, 0], resp_lowd[:, 1]
    pts = ax.scatter(x, y, c=time, cmap='twilight', vmin=0, vmax=2000)
    
  
    ax.set_xlabel('Dimension 1')
    ax.set_ylabel('Dimension 2')
    ax.set_xticks([])
    ax.set_yticks([])
        
  fig.colorbar(pts, ax=ax, ticks=np.linspace(0, 2000, 9), label='time')



roiTraj(11, 12, 8)


roiTraj(11, 30, 13)



```
